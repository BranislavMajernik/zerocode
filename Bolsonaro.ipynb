```python
import dash
from dash import dcc, Input, Output
from jupyter_dash import JupyterDash
import dash_table as dt
import pandas as pd
import dash_bootstrap_components as dbc

df = pd.read_csv('https://git.io/Juf1t')

external_stylesheets = [dbc.themes.BOOTSTRAP]

app = JupyterDash(external_stylesheets=external_stylesheets)

app.layout = dbc.Container([
    dbc.Label('Click a cell in the table:'),
    dt.DataTable(
        id='tbl', data=df.to_dict('records'),
        columns=[{"name": i, "id": i} for i in df.columns],
    ),
    dbc.Alert("Click the table", id='out'),
])

@app.callback(Output('out', 'children'), Input('tbl', 'active_cell'))
def update_graphs(active_cell):
  return str(active_cell)

app.run_server(mode='inline')
```

```python
# V state ConspiracyRepublic sa voli prezident z dvoch kandidatov. Napriklad ako v druhom kole v Brazili.
# Krajina ma pre jednoduchost 6 volebnych obvodov, v kazdom po 1000 volicov,
# ktory su povinny ist k volbam a teda pocitajme za kazdy obvod 1000 odovzdanych hlasov.
ob = np.array(["Obvod 1","Obvod 2","Obvod 3","Obvod 4","Obvod 5","Obvod 6"])
#Tu su vysledky pre kandidata Dezolatovic
y1 = np.array([700,500,300,400,450,300])
plot.plot(ob,y1, color='r', label="Dezolatovic")

#Tu su vysledky pre kandidata Slnieckovic
y2 = np.array([300,500,700,600,550,700])
plot.plot(ob,y2, color='b', label="Slnieckovic")

plot.xlabel('Volebne Obvody')
plot.ylabel('Pocty hlasov')
plot.legend(loc="center right")
```

```python
# Volebna komisia sa rozhodla zverejnovat priebezne ako aj celkove vysledky v percentach (Podobne ako Brazilia). Z tohoto dovodu urobila nasledovny vypocet:
cum1 = []
cum2 = []
sum1 = 0
sum2 = 0
i=0
j=0
k=0
l=0
for a in y1:
  i = i+1
  sum1 = sum1 + a
  j = i*1000
  print("Priebezny pocet hlasov pre kandidata Dezolatovica: " , sum1 , ".V percentach je to:", sum1*100/j, "%, pri celkovom pocte odovzdanych hlasov: ", j , ", co je v percentach:", j*100/6000, "%")
  cum1.append(sum1*100/j)
for b in y2:
  k= k+1
  sum2 = sum2 + b
  l = k*1000
  print("Priebezny pocet hlasov pre kandidata Slnieckovica: " , sum2,  ".V percentach je to:", sum2*100/l, "%, pri celkovom pocte odovzdanych hlasov: ", l , ", co je v percentach:", l*100/6000, "%")
  cum2.append(sum2*100/l)
```

```python
x = np.array([16.66,33.33,50,66.66,83.33,100])
plot.plot(x,cum1, color='r', label="Dezolatovic")
plot.plot(x,cum2, color='b', label="Slnieckovic")
plot.xlabel('Celkovy pocet scitanych hlasov v %')
plot.ylabel('Pocty hlasov v %')
plot.legend(loc="upper right")
```
